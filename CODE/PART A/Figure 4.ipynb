```python
%matplotlib inline
from scipy.integrate import odeint  
import numpy as np
from matplotlib import pyplot as plt
from shapely.geometry import LineString
```

```python
# Figure 4A

def sdot_toggle(s,t,params):
    # variables and parameters
    P1, P2 = s
    kp1, kp2, kdp1, kdp2, K1, K2, n1, n2 = params
    
    # rate equations of the toggle switch
    rate_P1_production = kp1 * ((K2**n2)/(P2**n2 + K2**n2))
    rate_P2_production = kp2 * ((K1**n1)/(P1**n1 + K1**n1))
    rate_P1_loss = -kdp1*P1
    rate_P2_loss = -kdp2*P2
    dP1 = rate_P1_production + rate_P1_loss
    dP2 = rate_P2_production + rate_P2_loss
    dsdt = [dP1, dP2]
    return dsdt

# Define parameter values and pack into 'params' tuple
# (rate constants use time units in minutes)
# cooperativity = 1
kp1 = 120.0
kp2 = 120.0
K1 = 3000.0
K2 = 3000.0
n1 = 1.0
n2 = 1.0
kdp1 = 1.0/60.0
kdp2 = 1.0/60.0
params = [kp1, kp2, kdp1, kdp2, K1, K2, n1, n2]

# Define list of observation points 't_obs'
t_start = 0
t_end = 1000
n_obs = 1000
t_obs = np.linspace(t_start, t_end, n_obs)

# Define initial conditions for high P1 over P2
# Use the odeint function to simulate system
# Unpack s1_obs (highP1/lowP2) into variable observations

P1 = 1000
P2 = 100
s1_0 = [P1, P2]

s1_obs = odeint(sdot_toggle, s1_0, t_obs, args=(params,))

P1_obs1 = s1_obs[:,0]
P2_obs1 = s1_obs[:,1]

# Define initial conditions for high P2 over P1
# Use the odeint function to simulate system
# Unpack s_obs (lowP1/highP2) into variable observations

P1 = 100
P2 = 1000
s2_0 = [P1, P2]

s2_obs = odeint(sdot_toggle, s2_0, t_obs, args=(params,))

P1_obs2 = s2_obs[:,0]
P2_obs2 = s2_obs[:,1]

# Plot results
fig3 = plt.figure(figsize=(10,6))
fig3.suptitle('Genetic toggle switch model for cooperativity of 2 (n1=n2=2)', fontsize=14, fontweight='bold')
axes1 = fig3.add_subplot(2,1,1)
axes2 = fig3.add_subplot(2,1,2)

axes1.plot(t_obs, P1_obs1, 'r-', label='P1')
axes1.plot(t_obs, P2_obs1, 'b-', label='P2')
axes2.plot(t_obs, P1_obs2, 'r-', label='P1')
axes2.plot(t_obs, P2_obs2, 'b-', label='P2')

axes2.set_xlabel('Time (min)')
axes1.set_ylabel('#Proteins copies')
axes2.set_ylabel('#Proteins copies')
axes1.xaxis.label.set_size(14)
axes1.yaxis.label.set_size(14)
axes2.xaxis.label.set_size(14)
axes2.yaxis.label.set_size(14)
axes1.xaxis.set_tick_params(labelsize=13)
axes1.yaxis.set_tick_params(labelsize=13)
axes2.xaxis.set_tick_params(labelsize=13)
axes2.yaxis.set_tick_params(labelsize=13)

axes2.legend(prop={'size': 15})

plt.savefig('fig4A.png');
```
