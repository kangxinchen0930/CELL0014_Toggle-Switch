```python
%matplotlib inline
from scipy.integrate import odeint  
import numpy as np
from matplotlib import pyplot as plt
```

```python
def sdot_repressor(s,t,params):
    
    # variables and parameters
    P, Pr = s
    kp, kdp, K, n = params
    
    # rate equations in the presence of repressor
    rate_P_production = kp * ((K**n)/(Pr**n + K**n))
    rate_P_loss = -kdp*P
    dP = rate_P_production + rate_P_loss
    dPr = 0
    dsdt = [dP, dPr]
    
    return dsdt
```

```python
# Define parameters' values
kp = 120.0
K = 3000.0
n = 4.0
kdp = 1.0/60.0
params = [kp, kdp, K, n]

# Define list of observation points 't_obs'
t_start = 0
t_end = 400
n_obs = 1000
t_obs = np.linspace(t_start, t_end, n_obs)
```

```python
# Figure 3

# For each cooperativity n value
    # Create a range of Pr values
    # Create an empty list to store the corresponding protein steady state (ss) values
    # Calculate the corresponding protein ss value for each repressor value and append the value to the empty list

Pr_vals = np.linspace(0,20000,100)

n = 1
P_vals_n1 = []
for Pr in Pr_vals:
    P_ss = (kp/kdp) * ((K**n)/(((K**n + Pr**n))))
    P_vals_n1.append(P_ss)

n = 2
P_vals_n2 = []
for Pr in Pr_vals:
    P_ss = (kp/kdp) * ((K**n)/(((K**n + Pr**n))))
    P_vals_n2.append(P_ss)
    
n = 4
P_vals_n4 = []
for Pr in Pr_vals:
    P_ss = (kp/kdp) * ((K**n)/(((K**n + Pr**n))))
    P_vals_n4.append(P_ss)
    
n = 8
P_vals_n8 = []
for Pr in Pr_vals:
    P_ss = (kp/kdp) * ((K**n)/(((K**n + Pr**n))))
    P_vals_n8.append(P_ss)
    
# Plot results
fig2 = plt.figure(figsize=(10,6))
fig2.suptitle('Protein level as a function of repressor level for different cooperativity values', fontsize=14, fontweight='bold')

ax = fig2.add_subplot(1,1,1)
ax.plot(Pr_vals, P_vals_n1, 'r-', label='n=1')
ax.plot(Pr_vals, P_vals_n2, 'b-', label='n=2')
ax.plot(Pr_vals, P_vals_n4, 'y-', label='n=4')
ax.plot(Pr_vals, P_vals_n8, 'g-', label='n=8')

ax.set_xlabel('#Repressor level')
ax.set_ylabel('#Protein level')
ax.xaxis.label.set_size(13)
ax.yaxis.label.set_size(13)
ax.xaxis.set_tick_params(labelsize=13)
ax.yaxis.set_tick_params(labelsize=13)
ax.legend(prop={'size': 15})

plt.savefig('fig3.png');
```
